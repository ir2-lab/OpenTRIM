```python
from pathlib import Path
import subprocess

workdir = Path("/home/yiannis/dimokritos/my-fork/OpenTRIM")  # FIXED (no ~/ inside)
exe = Path("/home/yiannis/.local/bin/opentrim")

cmd = [str(exe), "--help"]   # start with help to ensure it exits quickly

# Optional: verify paths early (so you see the real error immediately)
assert workdir.is_dir(), f"workdir not found: {workdir}"
assert exe.is_file(), f"exe not found: {exe}"

result = subprocess.run(
    cmd,
    cwd=workdir,
    text=True,
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
    timeout=20,   # prevents “endless loading”
)

print("returncode:", result.returncode)
print("stdout:\n", result.stdout)
print("stderr:\n", result.stderr)
```
